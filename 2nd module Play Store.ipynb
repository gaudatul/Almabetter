## Handling the NaN values

Installs, Size, and Price will be the three features we use the most frequently moving forward. A close examination of the dataset indicates that several of these columns require data cleansing in order to be used by later-written algorithms. The Installs, Size, and Price columns, in particular, are challenging to convert to a numerical data type due to the existence of special characters (, $ +) and letters (M k). Let's tidy up by getting rid of them and changing the column types to numbers.

```python
app['Rating'].isnull().sum()
```

There are 1474 NaN values in the 'Rating' column, or roughly 13.5% of all the rows in the dataset. It is not realistic to remove these rows because doing so would result in a significant loss of data and could affect the analysis's overall quality. The sum (mean or median) of the remaining values in the Rating column can be used to impute the NaN values in this situation.

First any value out of range between 1 and 5 in rating column needs to be checked

```python
app[(app['Rating'] < 1) | (app['Rating'] > 5)]
```

Since there is only one value out of the range of the Rating between 1 and 5, it can be dropped

```python
app.drop([10472], axis=0, inplace=True)
```

```python
#app.drop([10472], 0, inplace = True)
```

```python
# Taking the NaN values out when calculating the mean and median in the Rating column.

mean_rating = round(app[~app['Rating'].isnull()]['Rating'].mean(),4)

median_rating = app[~app['Rating'].isnull()]['Rating'].median()

[mean_rating , median_rating]
```

To decide filling NaN values of columns with mean or median we need to identify the outliers

```python
fig,ax = plt.subplots(2,1,figsize = (12,6))
sns.histplot(app['Rating'], kde = True, ax = ax[0])
sns.boxplot(data = app, x = 'Rating', ax = ax[1])
```

The mean is skewed by the outliers, and the median is 4.3. Consequently, the NaN values can be filled using the median.

```python
app['Rating'].fillna(value = median_rating, inplace = True)
```

```python
app['Rating'].isnull().sum()
```

## Handling the duplicated values

```python
app['App'].value_counts()
```

```python
print(f"The number of Duplicated apps is {app['App'].duplicated().sum()}")
```

```python
# Remove Duplicate Apps
app.drop_duplicates(subset = 'App', keep = 'first', inplace = True)
print(app.shape)
```

### Changing the Last Updated column's datatype from string to datetime.

```python
app["Last Updated"] = pd.to_datetime(app['Last Updated'])
app.head()
```

### Converting the datatype of price column to FLoat

```python
# Converts price column to float type.
app['Price'] = app['Price'].apply(lambda x: x.replace("$",""))
app['Price'] = app['Price'].astype(float)
```

### Converting the installs column to int datatype

```python
# Data Cleaning and Type Conversion of 'Installs'.
app['Installs'] = app['Installs'].str.replace("+","")
app['Installs'] = app['Installs'].str.replace(",","")
app['Installs'] = app['Installs'].astype(int)
```

### Handling size column

```python
app['Size'].unique()
```

```python
# Modifying the 'Size' Column in a DataFrame
app['Size'] = app['Size'].str.replace("k","")
app['Size'] = app['Size'].str.replace("M","*1000")
app['Size'] = app['Size'].str.replace("Varies with device","-1")
app['Size'] = app['Size'].apply(lambda x: eval(x))
app['Size'] = app['Size'].replace(-1,np.nan)
app['Size'].fillna(np.mean(app['Size']),inplace = True)
```

## **3. What are the top 6 categories and its rating distribution ?**

###Information about the Dataset
####User Review dataframe has 64295 rows and 5 columns. The 5 columns are identified as follows:

####App: Contains the name of the app with a short description (optional).
####Translated_Review: It contains the English translation of the review dropped by the user of the app.
####Sentiment: It gives the attitude/emotion of the writer. It can be ‘Positive’, ‘Negative’, or ‘Neutral’.
####Sentiment_Polarity: It gives the polarity of the review. Its range is [-1,1], where 1 means ‘Positive statement’ and -1 means a ‘Negative statement’.
####Sentiment_Subjectivity: This value gives how close a reviewers opinion is to the opinion of the general public. Its range is [0,1]. Higher the subjectivity, closer is the reviewers opinion to the opinion of the general public, and lower subjectivity indicates the review is more of a factual information.
